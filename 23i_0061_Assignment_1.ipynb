```python
class Agent:
    def __init__ (self, X,Y):
        self.X_cor = X
        self.Y_cor = Y
        self.percept = []
        self.cherries = 0
    def display_Position (self) :
        print (f'<-------------------( {self.X_cor} , {self.Y_cor} )------------------->')
    def display_Cherries (self) :
        print (f'<-------------------( Cherries : {self.cherries} )------------------->')
    def mov (self, direction):
        if (direction == 'left'):
            self.percept.append ((self.X_cor , self.Y_cor))
            self.Y_cor -= 1
        elif (direction == 'right'):
            self.percept.append ((self.X_cor , self.Y_cor))
            self.Y_cor += 1
        elif (direction == 'up'):
            self.percept.append ((self.X_cor , self.Y_cor))
            self.X_cor -= 1
        else:
            self.percept.append ((self.X_cor , self.Y_cor))
            self.X_cor += 1
    def isVisited (self, point):
        return point in self.percept
    def print_percept (self):
        print (self.percept)
    def isProtected (self):
        return self.cherries > 0

agent = Agent (0,0) 
```

```python
graph = {'Service Room':[('Room C', 5), ('Room B', 5), ('Room A', 1)],
        'Room A':[('Service Room', 0)],
        'Room B':[('Service Room', 0)],
        'Room C':[('Service Room', 0)]
        
        }

def DFS (graph , start ) :
    visited = {}
    cost = 0
    for k in graph.keys():
        visited[k] = False
    stack = []
    stack.append (start)
    

    while (stack):
        # print (stack)
        
        num = stack.pop()
        cost = cost + num[1]
        print (num, end = ' -> ')
        if (num[0] != 'Service Room'):
            visited[num[0]] = True
        
        
        for v in graph [num[0]]:
            if visited[v[0]] == False and v not in stack :
                stack.append (v)
        
    print ()
    print ('Total cost: ', cost, end = ' ')
    print ()   

print ('DFS')
DFS(graph, ('Service Room', 0))

    
```

```python
graph = {'Service Room':[('Room B', 5), ('Room A', 5), ('Room C', 1)],
        'Room A':[('Service Room', 0)],
        'Room B':[('Service Room', 0)],
        'Room C':[('Service Room', 0)]
        
        }

def IDDFS (graph , start, depth ) :

    for dep in range (depth) :
        print (f'<-------------------( Iteration {dep+1} )------------------->')
        visited = {}
        cost = 0
        for k in graph.keys():
            visited[k] = False
        stack = []
        
        stack.append ((start, 0))
        

        while (stack):
            # print (stack)
            
            num = stack.pop()
            cost = cost + num[0][1]
            print (num, end = ' -> ')
            if (num[0][0] != 'Service Room'):
                visited[num[0][0]] = True
            if (num[1] >= dep):
                continue
            temp_list = [s[0] for s in stack ]
            for v in graph [num[0][0]]:
                if visited[v[0]] == False and v not in temp_list :
                    if (v[0] == 'Service Room'):
                        stack.append ((v, 0))
                    else:
                        stack.append ((v, num[1]+1))
            
        print ()
        print ('Total cost: ', cost, end = ' ')
        print ()   

print ('IDFS')
IDDFS(graph, ('Service Room', 0), 3) # (()'Service room', cost), depth)
```
