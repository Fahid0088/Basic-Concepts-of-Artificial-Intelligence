```python
graph = {'A':['S'],
        'B':['C','D','S'],
        'C':['B', 'J'],
        'D':['B','S','G'],
        'E':['S','G'],
        'F':['H','G'],
        'G':['D','E','F','H','J'],
        'H':['F', 'G', 'I'],
        'I':['H', 'J'],
        'J':['C', 'G', 'I'],
        'S':['A','B','D', 'E']
        }

def DLS (graph, start, goal, depth):
    dict={}
    for key in graph.keys():
        dict [key] = ['&', False]
    print (dict)
    stack = []
    stack.append ( (start, 0))
    while (len (stack) > 0):
        parent, dep = stack.pop()
        if dep == depth :
            continue
        print (parent)
        dict[parent][1]= True
        if (parent == goal):
            current = goal
            path = []
            while current != '&':
                path.append(current)
                current = dict [current][0]
            print (path[::-1])
            print ('goal found')
            return
        for child in graph [parent]:
            if dict[child][1] == False and child not in stack:
                dict[child][0] = parent
                stack.append ((child, dep + 1))
    print ('goal not found')

DLS(graph, 'S', 'I', 5)
```

```python
def IDFS (graph, start, goal, depth):
    for d in range (1,depth+1):
        print ('----------------------------------------------------------')
        DLS(graph, start, goal, d)

IDFS(graph, 'S', 'I', 5)
```
