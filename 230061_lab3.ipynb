```python
location = input ('Enter the current location of the Vaccume: ')
list = ['A', 'B', 'C']
check_S = bool( int (input (f'Enter the status of room {location} (1/0)')))
print (check_S)
check = []
room = []
check.append (check_S)
room.append (location)

for a in list:
    if a != location:
        check_O1 = bool (int(input (f'Enter the status of room {a} (1/0)')))
        
        check.append (check_O1)
        room.append (a)
print (check)
current_location = location
cost = 0
for a in range (len (check)):
    if check[a]:
        if (current_location == room[a]):
            print (f'Room {room[a]} is dirty......... . . .  .  .')
            print (f'Agent is sucking the Room {room[a]} (adding 1 to cost) ......... . . .  .  .')
            cost += 1
            check[a] = 0
        else:
            print (f'Room {room[a]} is dirty........ . . . .  .  .  .')
            print (f'Agent is moving from Room {current_location} to Room {room [a]} (adding 1 to cost) ........ . . . .  .  .  .')
            cost += 1
            current_location = room [a]
            print (f'Agent is sucking the Room {room[a]} (adding 1 to cost) ......... . . .  .  .')
            cost += 1
            check[a] = 0


print ('Agent have achieved the goal all rooms are clean now. ')  
print (f'Total cost to do so is: {cost}')
            
        
```
